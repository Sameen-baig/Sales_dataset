```python
# changing the type of 'Date' & 'Time' column - 
df['Date']= pd.to_datetime(df['Date'])

# Convert the 'time_column' to datetime
df['Time'] = df['Time'].astype(str)
df['Time'] = pd.to_datetime(df['Time'])
df['Time'] = df['Time'].dt.strftime('%H:%M')

#Creating two new columns from 'Date' column -
df['month'] = df['Date'].dt.month
df['year'] = df['Date'].dt.year
df['Day'] = df['Date'].dt.day_name()

df.sample(3)

# converting the 'Date' column in str format-
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
```

```python
# removing space from the column names and convertin them to lowercase-
df.columns = df.columns.str.replace('', '').str.lower()
```

# EDA

```python
# Sales by Gender
sales_by_gender = df.groupby('gender')['total'].sum()
sales_by_gender.plot(kind='bar', title='Total Sales by Gender', color=['blue', 'pink'])
plt.ylabel('Total Sales')
plt.show()
```

```python
#Sales by Product 
sales_by_category = df.groupby('product line')['total'].sum()
sales_by_category.plot(kind='bar', title='Total Sales by Product Category', color='green')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Spending Behavior Analysis
avg_spending_by_gender = df.groupby('gender')['total'].mean()
avg_spending_by_gender.plot(kind='bar', title='Average Spending by Gender', color=['blue', 'pink'])
plt.ylabel('Average Spending')
plt.show()
```

```python
# Sales by Payment Method
sales_by_payment = df.groupby('payment')['total'].sum()
sales_by_payment.plot(kind='pie', title='Sales by Payment Method', autopct='%1.1f%%', figsize=(4, 2))
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
colormap = sns.color_palette("PiYG")
sns.heatmap(df.drop("gross margin percentage", axis = 1).corr(numeric_only=True), annot=True, cmap=colormap).set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12)
plt.show()
```

```python
# Total Sales vs. Branches
sns.barplot(data=df, x='branch', y='total', ci=None)
plt.title('Total Sales vs. Branches')
plt.show()
```

```python
# Sales vs. Payment Method
sns.boxplot(data=df, x='payment', y='total')
plt.title('Sales Distribution by Payment Method')
plt.show()

# Ratings vs. Sales
sns.scatterplot(data=df, x='rating', y='total', hue='branch', palette='viridis')
plt.title('Ratings vs. Sales')
plt.xlabel('Customer Ratings')
plt.ylabel('Total Sales')
plt.show()
```

# Visualization

```python
## 1. Pie Chart: Sales Distribution by Branch ###
sales_by_branch = df.groupby('branch')['total'].sum()
sales_by_branch.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(4, 4), colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Sales Distribution by Branch')
plt.ylabel('')  # Hide the y-label for aesthetics
plt.show()
```

```python
### 2. Bar Chart: Payment Methods vs. Total Sales ###
sales_by_payment = df.groupby('payment')['total'].sum()
sales_by_payment.plot(kind='bar', color=['#ff9999', '#66b3ff', '#99ff99'], figsize=(4, 4))
plt.title('Payment Methods vs. Total Sales')
plt.xlabel('Payment Method')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.show()
```

```python
### 4. Boxplot: Sales Distribution by Customer Gender ###
plt.figure(figsize=(4, 2))
sns.boxplot(data=df, x='gender', y='total', palette='Set2')
plt.title('Sales Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales')
plt.show()
```
